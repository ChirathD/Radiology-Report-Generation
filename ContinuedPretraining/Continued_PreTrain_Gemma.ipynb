```python
import pandas as pd
from datasets import load_dataset
import datetime

import torch
from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, GemmaTokenizer
```

```python
from peft import LoraConfig

lora_config = LoraConfig(
    r=8,
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj", "gate_proj", "up_proj", "down_proj"],
    task_type="CAUSAL_LM",
)
```

```python
model_id = "google/gemma-2b"
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    bnb_8bit_quant_type="nf4",
    bnb_8bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained('/content/Checkpoint-19500', token="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")
model = AutoModelForCausalLM.from_pretrained('/content/Checkpoint-19500', quantization_config=bnb_config, device_map={"":0}, token="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")
```

```python
dataset = load_dataset('LexiconShiftInnovations/SinhalaCorpusSubset')
```

```python
import transformers
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset['train'],
    dataset_text_field="Text",
    args=transformers.TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=1,
        warmup_steps=2,
        max_steps=20000,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit",
        report_to="wandb",
        run_name=f"Gemma2bsinhala-{datetime.datetime.now().strftime('%Y-%m-%d-%H-%M')}",
    ),
    peft_config=lora_config,
)
```

```python
trainer.train()
```

```python
model_id = "google/gemma-2b"
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    bnb_8bit_quant_type="nf4",
    bnb_8bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained('/content/checkpoint-18000', token="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")
model = AutoModelForCausalLM.from_pretrained('/content/checkpoint-18000', quantization_config=bnb_config, device_map={"":0}, token="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")
```

```python
txt = 'කුහර ඇතිවීම වැළැක්වීම සඳහා මා වැළකී සිටිය යුතු ආහාර සහ බීම මොනවාද?'
text = f"""Instruction:\n{txt}\nResponse:"""

inputs = tokenizer(text, return_tensors="pt")

outputs = model.generate(**inputs, max_new_tokens=100, do_sample=True, temperature=0.1)
print(tokenizer.decode(outputs[0],skip_special_tokens=True))
```

```python
from peft import LoraConfig

lora_config = LoraConfig(
    r=8,
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj", "gate_proj", "up_proj", "down_proj"],
    task_type="CAUSAL_LM",
)
```

```python
import transformers
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    dataset_text_field="text",
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=1,
        warmup_steps=2,
        max_steps=1000,
        learning_rate=2e-5,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit",
        report_to="wandb",
        run_name=f"Gemma2bsinhala-{datetime.datetime.now().strftime('%Y-%m-%d-%H-%M')}",
    ),
    peft_config=lora_config,
)
```

```python
trainer.train()
```
